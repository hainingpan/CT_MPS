# API

```python
def plot_T(data, p, metric, method, rescale=False, idxmin=1, xscale='log',yscale='log', xlim=None, z = 2,     beta = 1, nu = 1):
    fig, ax = plt.subplots(figsize=(6,3))
    color_list = plt.cm.Blues(np.linspace(0.3,1,4))
    

    for idx, L in enumerate([10,20,30,40]):
        x = np.arange(len(data[p,L]))[idxmin:]
        y = data[p,L][idxmin:]
        if rescale:
            k = 2 if metric=='var' else 1
            ax.plot(x/L**z, y*L**(k*beta/nu), label=f'L={L}', color=color_list[idx])
        else:
            ax.plot(x, y, label=f'L={L}', color=color_list[idx])
    if yscale:
        ax.set_yscale(yscale)
    if xscale:
        ax.set_xscale(xscale)
    sigma2 = rf'\sigma_{{{method}}}^2'
    yscale = {'weight':r'$L^{\beta/\nu}$', 'mean':r'$L^{\beta/\nu}$', 'var': r'$L^{2\beta/\nu}$'}
    name = {'weight': rf'$P({sigma2}=0)$', 'mean': rf'Mean$({sigma2})$', 'var': rf'Var$({sigma2})$'}


    if rescale:
        ax.set_xlabel(r'$t/L^z$')
        ax.set_ylabel(name[metric] + yscale[metric])
    else:
        ax.set_xlabel('$t$')
        ax.set_ylabel(name[metric])
    # ax.set_ylim(1e-5,None)
    if xlim:
        ax.set_xlim(xlim)
```
